In this assignment, we analyze the performance of nonlinear Support Vector Machine (SVM) classifiers using different kernel functions. In particular, we compare polynomial (poly) and radial basis function (RBF) kernels and study how their performance depends on hyperparameter choices.

The analysis is conducted on a star classification dataset, where the goal is to predict the type of a star based on its physical properties. The dataset includes features such as temperature, luminosity, radius, and absolute magnitude, which are derived from astrophysical observations. The target variable is the star type, representing different categories such as dwarfs, giants, and supergiants. This results in a multi-class classification problem.

Unlike many engineered tabular datasets, the features in this dataset are grounded in real physical relationships. However, these relationships are highly nonlinear (e.g., the relation between temperature and luminosity), making this dataset particularly suitable for kernel-based methods such as SVM. Linear models are expected to struggle, while nonlinear kernels should better capture the underlying structure.

We evaluate model performance using two metrics: accuracy and macro-averaged F1-score. Accuracy provides an overall measure of correctness, while the F1-score accounts for class imbalance and evaluates performance across all classes more robustly.

The objective of this assignment is to determine which kernel and hyperparameter combination yields the best performance and to analyze how model complexity affects classification results.
